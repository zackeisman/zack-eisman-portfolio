# Side by Side Boxplots

```r
library(tidyverse)
```

```r
# Reading in velocity csv file
velocity <- read.csv("ball_velocity_with_success.csv")

head(velocity, n = 10)
```

```r
# Boxplot of Catcher Throw Velocity
boxplot(velocity$velocity_m_per_s,
        main = 'Boxplot of Catcher Velocity',
        ylab = 'Velocity (m/s)',
        outline = FALSE)
```

```r
#Grouped Boxplot showing difference in throw velocity for safe and out runners

velocity_boxplot <- boxplot(velocity$velocity_m_per_s ~ velocity$success,
        main = "Velocity Boxplots by Result",
        ylab = "Velocity (in m/s)",
        xlab = "Safe vs Out",
        names = c("safe","out"),
        col = c("dark green", "red"),
        outline = FALSE
)
```

```r
#Numerical summaries for safe vs out velocities
aggregate(velocity_m_per_s ~ success,
          data = velocity,
          FUN = mean
          )
```

```r
# Reading in velocity csv file
pop_time <- read.csv("pop_time_with_success_and_velocity.csv")

head(pop_time, n = 10)
```

```r
# Boxplot of Catcher Pop Time
boxplot(pop_time$pop_time,
        main = 'Boxplot of Catcher Pop Time',
        ylab = 'Pop Time (seconds)',
        outline = FALSE
        )
```

```r
#Grouped Box Plot showing difference in pop time for safe and out runners

boxplot(pop_time$pop_time ~ pop_time$success,
        main = "Pop Time Boxplots by Result",
        ylab = "Pop Time (in seconds)",
        xlab = "Safe vs Out",
        names = c("safe","out"),
        col = c("dark green", "red"),
        outline = FALSE
)
```

```r
#Numerical summaries for safe vs out pop times
aggregate(pop_time ~ success,
          data = pop_time,
          FUN = mean
          )
```

```r
# Reading in leadoff csv file
leadoff <- read.csv("first_base_runner_distance_with_success.csv")

head(leadoff, n = 10)
```

```r
# Boxplot of Runner's Leadoff
boxplot(leadoff$distance_from_1st_base_ft,
        main = 'Boxplot of Runner Leadoff',
        ylab = 'Leadoff (feet from 1st base)',
        outline = FALSE)
```

```r
#Grouped Box Plot showing difference in leadoff for safe and out runners

boxplot(leadoff$distance_from_1st_base_ft ~ leadoff$success,
        main = "Leadoff Boxplots by Result",
        ylab = "Leadoff (in feet)",
        xlab = "Safe vs Out",
        names = c("safe","out"),
        col = c("dark green", "red"),
        outline = FALSE
)
```

```r
#Numerical summaries for safe vs out leadoffs
aggregate(distance_from_1st_base_ft ~ success,
          data = leadoff,
          FUN = mean
          )
```
